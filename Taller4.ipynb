```python
import math
from random import random
import matplotlib.pyplot as plt
import numpy as np


def distancia(x, y):
  return math.sqrt(x**2 + y**2)
```

```python
cont = 0
cantPuntos = 1000

i = 0
while i < cantPuntos:
  x = random()
  y = random()
  i = i+1
  if distancia(x, y) <= 1:
      cont += 1
      plt.plot(x, y, 'ro')
  else:
      plt.plot(x, y, 'bo')

print("Puntos dentro de la circunferencia:", cont, "/", cantPuntos)
print("Estimación de pi:", (cont * 4 / (cantPuntos)))

#Grafica Linea de separación
x = np.arange(-1, 1, 0.01)
y = (1 - x**2)**(1/2)
plt.plot(x,y)

#Ajuste Ejes
plt.axis([0, 1.05, 0, 1.05])

plt.show()
```
